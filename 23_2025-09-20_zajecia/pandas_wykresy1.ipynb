# Wykres pudekowy (box plot)

Wykres pudekowy to graficzna prezentacja rozkadu danych jednowymiarowych, opierajca si na piciu kluczowych wartociach:

- **Minimum (Min)**  najmniejsza warto danych,
- **Pierwszy kwartyl (Q1)**  warto, poniej ktrej znajduje si 25% danych,
- **Mediana (Me)**  rodkowa warto danych dzielca je na poowy,
- **Trzeci kwartyl (Q3)**  warto, poniej ktrej znajduje si 75% danych,
- **Maksimum (Max)**  najwiksza warto danych.

## Interpretacja poszczeglnych elementw wykresu:

- **Pudeko**: rozciga si od Q1 do Q3 i obejmuje 50% rodkowych danych. Dugo pudeka to rozstp midzykwartylowy (IQR), czyli Q3  Q1.
- **Linia w rodku pudeka**: oznacza median  warto rodkow rozkadu.
- **Wsy**: linie rozcigajce si od pudeka do wartoci minimalnej i maksymalnej (lub do wartoci granicznych definiowanych przez 1,5  IQR). Pokazuj rozpito typowych danych.
- **Punkty poza wsami (outliers)**: wartoci ekstremalne, ktre znacznie odbiegaj od pozostaych obserwacji.

## Co mona odczyta z wykresu pudekowego?

- Rozkad danych i ich rozproszenie (im duszy wykres, tym wiksza zmienno).
- Pooenie mediany i symetri rozkadu (jeli mediana jest blisko rodka pudeka, rozkad jest symetryczny).
- Obecno wartoci odstajcych.
- Ewentualn asymetri danych, np. wyduone wsy z jednej strony wskazuj na prawostronn lub lewostronn asymetri.

Wykres pudekowy jest powszechnie uywany do porwnywania rozkadw w rnych grupach, zwaszcza przy analizie statystycznej i eksploracji danych.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Zamy, e df ju jest zdefiniowany

# Tworzymy przedziay co 50 kalorii
bins = range(0, int(df['Calories'].max()) + 50, 50)
df['Calories_bin'] = pd.cut(df['Calories'], bins=bins, right=False)

g = sns.catplot(
    data=df,
    x="Calories_bin", y="Pulse",
    kind="box",
    height=5,
    aspect=2
)

plt.xticks(rotation=45)  # Obrcenie etykiet do czytelnoci
plt.show()
```
